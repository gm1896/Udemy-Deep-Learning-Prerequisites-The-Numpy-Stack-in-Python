# Line Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,10)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Some Function of Time")
plt.title("Cool Chart")
plt.show()
```

```python
x1 = np.linspace(0,10,100)
y1 = np.sin(x1)
plt.plot(x1,y1)
plt.xlabel("Time")
plt.ylabel("Some Function of Time")
plt.title("Cool Chart")
plt.show()
```

# Scatter Plot

```python
A = pd.read_csv("data_1d.csv",header=None).values
```

```python
x = A[:,0]
y = A[:,1]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_line =  np.linspace(0,100,100)
y_line = 2*x_line +1
```

```python
plt.scatter(x,y)
plt.plot(x_line,y_line)
plt.show()
```

# Histogram
